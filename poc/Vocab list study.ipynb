# read a corpous, extract word, transform stem, and count frequency

```python
f = open("./data/english/corpos/harry_5_1.txt", "r", encoding='UTF-8')

from collections import defaultdict
ps = PorterStemmer()
wordnet_lemmatizer = WordNetLemmatizer()


def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wn.ADJ,
                "N": wn.NOUN,
                "V": wn.VERB,
                "R": wn.ADV}
    return tag_dict.get(tag, wn.NOUN)
    
corposWord = []
# read
for line in f:
    words = word_tokenize(line)

    for w in words:
        corposWord.append(wordnet_lemmatizer.lemmatize(w , get_wordnet_pos(w)))
```

```python
def wordListToFreqDict(wordlist):
    wordfreq = [wordlist.count(p) for p in wordlist]
    return dict(list(zip(wordlist,wordfreq)))

def sortFreqDict(freqdict):
    aux = [(freqdict[key], key) for key in freqdict]
    aux.sort()
    aux.reverse()
    return aux

dictionary = wordListToFreqDict(corposWord)
sorteddict = sortFreqDict(dictionary)

for s in sorteddict: print(str(s))
```

# The whole process

```python
def processCorpos(file):
    f = open(file, "r", encoding='UTF-8')
    
    corposWord = []
    # read
    for line in f:
        words = word_tokenize(line)

        for w in words:
            lw = w.lower()
            corposWord.append(wordnet_lemmatizer.lemmatize(lw , get_wordnet_pos(lw)))
        
    corposWord2 = removeStopwords(corposWord, stopwords)
    dictionary = wordListToFreqDict(corposWord2)
    sorteddict = sortFreqDict(dictionary)
    
    # the counting
    a1_cnt= 0
    a1_word = set()
    a2_cnt= 0
    a2_word = set()
    b1_cnt= 0
    b1_word = set()
    b2_cnt= 0
    b2_word = set()
    c1_cnt= 0
    c1_word = set()
    c2_cnt= 0
    c2_word = set()
    unc_word =set()

    for s in sorteddict:
        if s[1] in a1:
            a1_cnt = a1_cnt+1
            a1_word.add(s[1])
        elif s[1] in a2:
            a2_cnt = a2_cnt+1
            a2_word.add(s[1])
        elif s[1] in b1:
            b1_cnt = b1_cnt+1
            b1_word.add(s[1])
        elif s[1] in b2:
            b2_cnt = b2_cnt+1
            b2_word.add(s[1])
        elif s[1] in c1:
            c1_cnt = c1_cnt+1
            c1_word.add(s[1])
        elif s[1] in c2:
            c2_cnt = c2_cnt+1
            c2_word.add(s[1])
        else:
            unc_word.add(s[1])
    
    return (a1_word, a2_word, b1_word, b2_word, c1_word, c2_word, unc_word)
```

```python
(a1w, a2w, b1w, b2w, c1w, c2w, uw) = processCorpos("./data/english/corpos/steve.txt")
```
